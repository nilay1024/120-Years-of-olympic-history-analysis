```python
# Hypothesis Testing
# The null hypothesis is: Mean height of athletes in Basketball winning a Gold is 193.
#    H0 : μ <= 105

# The alternate hypothesis is: Average mean of athletes in Basketball winning a Gold is NOT 193.
#    H1 : μ > 105
```

```python
import math
arr = list()
for i in range(len(df1["Height"])):
    if (df1["Sport"][i]) == "Basketball" and (df1["Medal"][i]) == "Gold":
        arr.append(df1["Height"][i])
# print(sum(arr)/len(arr))
population_mean = sum(arr)/len(arr)
print(population_mean)
index = np.random.choice(arr,100)
sample_mean = math.ceil(np.mean(index))
print(sample_mean)
```

```python
arr1 = np.array(arr)
b = arr1.std()
print(math.ceil(b))
```

```python
from scipy.stats import norm
from math import sqrt
def two_sided_hypo(sample_mean, pop_mean, std_dev, sample_size, alpha):
    actual_z = abs(norm.ppf(alpha/2))
    hypo_z = (pop_mean - sample_mean) / (std_dev/sqrt(sample_size))
    print('Actual z value :', actual_z)
    print('Hypothesis z value :', hypo_z, '\n')
    if hypo_z >= actual_z or hypo_z <= -(actual_z):
        return True
    else:
        return False
    

alpha = 0.05
pop_mean = population_mean
sample_size =  100
std_dev = 11

print('H0 : μ ==', sample_mean)
print('H1 : μ !=', sample_mean)
print('The value of alpha is :', alpha, '\n')

reject = two_sided_hypo(sample_mean, pop_mean, std_dev, sample_size, alpha)
if reject:
    print('Null hypothesis rejected.')
    print('Mean height of athletes in Basketball winning a Gold is not',sample_mean, '.')
else:
    print('Failed to reject null hypothesis.')
    print('Hence it is plausible that mean height of athletes in Basketball winning a Gold is', sample_mean, '.')

```
